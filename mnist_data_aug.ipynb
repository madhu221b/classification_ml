```python
def shift_image(image, dy, dx):
    some_digit_image = image.reshape(28, 28)
    shifted_image = shift(some_digit_image, [dy,dx], cval=0, mode="constant")
    return shifted_image.ravel()
```

```python
for dy, dx in ((1,0), (-1,0), (0,1), (0,-1)):
    for image, label in zip(X_train_augmented, y_train_augmented):
        new_image = shift_image(image, dy, dx)
        X_train_augmented = np.vstack([X_train_augmented,new_image])
        y_train_augmented = np.append(y_train_augmented,label)
```
