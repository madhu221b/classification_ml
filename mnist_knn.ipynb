```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
```

```python
X, y = mnist["data"], mnist["target"]
```

```python
"""Splitting into training & testing data"""
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
"""knn classifier"""
from sklearn.neighbors import KNeighborsClassifier
nclf = KNeighborsClassifier()
```

```python
nclf.fit(X_train, y_train)
```

```python
pred = nclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
```

```python
acc
```

```python
"""Grid search CV"""
from sklearn.model_selection import GridSearchCV
import numpy as np
params = {
    "weights": ("uniform", "distance"),
    "n_neighbors": np.arange(3, 100, 2) 
}
```

```python
nclf_2 =  KNeighborsClassifier()
```

```python
grid_search = GridSearchCV(nclf_2, params, cv=5,
                          scoring="accuracy", return_train_score=False)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
```

```python
cvres
```

```python
final_model = grid_search.best_estimator_
```

```python
final_predictions = final_model.predict(X_test)
```

```python
acc = accuracy_score(y_test, final_predictions)
```
